```python
import pandas as pd

# Load raw data
df_final = pd.read_csv("../data/softball_data.csv")

# Load schools
schools = [
    {"school": "california-lutheran", "enrollment": 2297, "city": "Thousand Oaks", "state": "CA", "lat": 34.226, "lon": -118.879},
    {"school": "redlands", "enrollment": 2045, "city": "Redlands", "state": "CA", "lat": 34.063, "lon": -117.164},
    {"school": "chapman", "enrollment": 7713, "city": "Orange", "state": "CA", "lat": 33.793, "lon": -117.851},
    {"school": "pomona-pitzer", "enrollment": 3008, "city": "Claremont", "state": "CA", "lat": 34.097, "lon": -117.711},
    {"school": "whittier", "enrollment": 767, "city": "Whittier", "state": "CA", "lat": 33.977, "lon": -118.032},
    {"school": "occidental", "enrollment": 1881, "city": "Los Angeles", "state": "CA", "lat": 34.126, "lon": -118.213},
    {"school": "la-verne", "enrollment": 3050, "city": "La Verne", "state": "CA", "lat": 34.100, "lon": -117.771}
]
df_schools = pd.DataFrame(schools)

# Merge with schools
df_merged = df_final.merge(
    df_schools,
    how="left",
    left_on="team",
    right_on="school"
)

# IMPORTANT: Drop rows with missing outcomes, but KEEP the year column intact
df_clean = df_merged[df_merged["outcome"].notna()].copy()

# Only convert date to datetime if it's not NaN - don't overwrite year
df_clean["date"] = pd.to_datetime(df_clean["date"], errors="coerce")

# FIX: Don't overwrite year! Year is already clean from scraping
# Just keep it as is

# Check for any remaining NaN years and fill from date if possible
df_clean["year"] = df_clean["year"].fillna(df_clean["date"].dt.year)

# Drop any rows still missing year
df_clean = df_clean[df_clean["year"].notna()].copy()

# Convert year to int
df_clean["year"] = df_clean["year"].astype(int)

# Save
df_clean.to_csv("data/clean_softball_data.csv", index=False)

print("Fixed! Chapman 2024:", len(df_clean[(df_clean["team"]=="chapman") & (df_clean["year"]==2024)]))
print("Occidental total:", len(df_clean[df_clean["team"]=="occidental"]))
```
